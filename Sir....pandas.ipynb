```python
import pandas as pd

data = {
    "Name": ["Amit", "Ravi", "Neha", "Pooja", "Ankit"],
    "Maths": [78, 56, 89, None, 67],
    "Science": [72, 60, None, 88, 70],
    "English": [65, 58, 75, 80, None]
}

df = pd.DataFrame(data)

# Fill missing values
df.fillna(df.mean(numeric_only=True), inplace=True)

# Add Total & Percentage
df["Total"] = df[["Maths", "Science", "English"]].sum(axis=1)
df["Percentage"] = df["Total"] / 3

# Top student
topper = df.loc[df["Percentage"].idxmax()]

print(df)
print("\nTopper:\n", topper)
```

```python
import pandas as pd

Data = {
    "Name": ["Amit","Ravi","Neha","Pooja","Ankit"],
    "Maths": [78,56,89,None,67],
    "Science": [72,60,None,88,70],
    "English": [80,81,82,83,84],
    "Hindi": [90,90,90,90,None]
}

df = pd.DataFrame(Data)
df.fillna(df.mean(numeric_only=True), inplace=True)
df["Total"] = df[["Maths","Science","English","Hindi"]].sum(axis=1)
df["Percentage"] = df["Total"]/3
topper = df.loc[df["Percentage"].idxmax()]
print(df)
print("\nTopper:\n",topper)
```

```python
data = {
    "Emp_ID": [101, 102, 103, 104, 105],
    "Department": ["IT", "HR", "IT", "Sales", "HR"],
    "Salary": [50000, 40000, 60000, 45000, None],
    "Experience": [2, 3, 5, 4, 1]
}
df = pd.DataFrame(data)
df["Salary"] = df.groupby("Department")["Salary"].transform(
    lambda x: x.fillna(x.mean())
)
print(df.groupby("Department")["Salary"].mean())
print(df[df["Salary"] > 50000])
print(df.sort_values(by="Salary", ascending=False))
```
