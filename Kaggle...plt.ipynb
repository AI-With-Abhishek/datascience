#pie chart

```python
import matplotlib.pyplot as plt
```

```python
Area=['Marketing','Sales','Development','HR','Customer']
Budget=[2.4,3.5,10,2,5.4]
```

```python
plt.pie(Budget,labels=Area)
plt.title("pie chart")
plt.show()
```

```python
Area=['Marketing','Sales','Development','HR','Customr']
Budget=[2.4,3.5,10,2.5,4]
colors=['green','red','blue','pink','yellow']
plt.pie(Budget,labels=Area,colors=colors,autopct='%0.1f%%',shadow=True,frame=True)
plt.show()
```

#scatter plot

```python
x = [10,15,20,25,30,35,40,50,70,64]
y = [45,34,67,87,55,22,89,90,11,88]
```

```python
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter graph')
plt.show()
```

```python
plt.scatter(x,y,color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter graph')
plt.show()
```

```python
plt.scatter(x,y,color='r',marker='D')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter graph')
plt.show()
```
